## 2\. Números

Preencha as células de código para preencher os valores de (A), (B) e (C) na tabela de ticket médio abaixo:

<br>

| Dia   | Valor Total Vendas | Qtd Total Vendas | Ticket Medio |
|-------|--------------------|------------------|--------------|
| 19/01 | (A)                | 3                | 320.52       |
| 20/01 | 834.47             | (B)              | 119.21       |
| 23/01 | 15378.12           | 5                | (C)          |

```
# (A)
qtde_total_vendasA = 3
ticket_medioA = 320.52

valor_total_vendasA = ticket_medioA / qtde_total_vendasA
print(valor_total_vendasA) 
```

```
# (B)
valor_total_vendasB = 834.47
ticket_medioB = 119.21

qtde_total_vendasB = valor_total_vendasB / ticket_medioB
print(int(qtde_total_vendasB))
```

```
# (C)
valor_total_vendasC = 15378.12
qtde_total_vendasC = 5

ticket_medioC = valor_total_vendasC / qtde_total_vendasC
print(ticket_medioC)
```

## 3\. Strings

Aplique três métodos distintos na *string* abaixo, você pode conferir alguns métodos neste [link](https://www.w3schools.com/python/python_ref_string.asp):

```
cancao = 'Roda mundo, roda gigante, roda moinho, roda pião.'

# deixa toda a string maiúscula
cancao.upper()
```

```
# substitui parte da string por outra palavra
cancao.replace('mundo','bola')
```

```
# procura string e retorna índice que está
cancao.find('pião')
```

Extraia da string abaixo o valor da taxa **selic** na variável `selic` e o valor do **ano** na variavel `ano`. Imprima os valores na tela.

```
noticia = 'Selic vai a 2,75% e supera expectativas; é a primeira alta em 6 anos.'
```

```
# extraindo Selic
selic = noticia[12:17]
print(selic)

#extraindo Ano
ano = noticia[-7]
print(ano + ' anos')
```
